```python
# Importando as Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Visualização do Dataset
df=pd.read_csv("spotify.csv")
print(df.head(10))
print(df.info())
print(df.describe())
```

```python
# Data Cleaning
df['acousticness'] = df['acousticness'].str.strip('kg')
df['danceability'] = df['danceability'].str.strip('mol/L')
df['loudness'] = pd.to_numeric(df['loudness'], errors='coerce')
df['song_duration_ms'] = pd.to_numeric(df['song_duration_ms'], errors='coerce')
df['speechiness'] = pd.to_numeric(df['speechiness'], errors='coerce')
df['liveness'] = pd.to_numeric(df['liveness'], errors='coerce')
df['audio_valence'] = pd.to_numeric(df['audio_valence'], errors='coerce')
df['tempo'] = pd.to_numeric(df['tempo'], errors='coerce')
df['instrumentalness'] = pd.to_numeric(df['instrumentalness'], errors='coerce')
df['energy'] = pd.to_numeric(df['energy'], errors='coerce')
df['acousticness'] = pd.to_numeric(df['acousticness'], errors='coerce')
df['time_signature'] = pd.to_numeric(df['time_signature'], errors='coerce')
df['song_popularity'] = pd.to_numeric(df['song_popularity'], errors='coerce')
df['key'] = df['key'].astype('category')
df['audio_mode'] = df['audio_mode'].astype('category')
df['danceability'] = pd.to_numeric(df['danceability'], errors='coerce')
df.info()
```

```python
# Análise de Distribuição
plt.subplot(1, 2, 1)
df['acousticness'].hist(bins=30, color='green', alpha=1, grid=False)
plt.title(label="Acousticness Distribution", fontsize=10)
plt.subplot(1, 2, 2)
df['danceability'].hist(bins=30, color='blue', alpha=1, grid=False)
plt.title(label="Danceability Distribution", fontsize=10)
plt.show()
plt.subplot(1, 2, 1)
df['energy'].hist(bins=30, color='#232164', alpha=1, grid=False)
plt.title(label="Energy Distribution", fontsize=10)
plt.subplot(1, 2, 2)
df['loudness'].hist(bins=30, color='purple', alpha=1, grid=False)
plt.title(label="Loudness Distribution", fontsize=10)
plt.show()
plt.subplot(1, 2, 1)
df['tempo'].hist(bins=30, color='violet', alpha=1, grid=False)
plt.title(label="Tempo Distribution", fontsize=10)
plt.subplot(1, 2, 2)
df['audio_valence'].hist(bins=30, color='pink', alpha=1, grid=False)
plt.title(label="Audio Valence Distribution", fontsize=10)
plt.show()
```

```python
# Análise de Relações entre variáveis
matriz_correlacao = df.corr()
mask = np.array(matriz_correlacao)
mask[np.tril_indices_from(mask)] = False
fig = plt.gcf()
fig.set_size_inches(30, 12)
sns.heatmap(data=matriz_correlacao, square=True, annot=True, cbar=True, cmap='BuPu', mask=mask)
plt.show()
sns.jointplot(x='loudness', y='energy', data=df, kind='kde', color='purple')
plt.show()
sns.jointplot(x='song_duration_ms', y='speechiness', data=df, kind='kde', color='purple')
plt.show()
```
